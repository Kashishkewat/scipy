# **converison of temperature**

```python
from scipy.constants import convert_temperature
temperature_in_fahrenheit=90
temperature_in_celsius=convert_temperature(temperature_in_fahrenheit, 'fahrenheit','celsius')
print(f'temparature in celsius:{temperature_in_celsius}')
```
